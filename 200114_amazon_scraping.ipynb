```python
import sys
#html操作
import requests
from bs4 import BeautifulSoup
#pandas
import pandas as pd
#gss
import gspread
import json
from oauth2client.service_account import ServiceAccountCredentials

#「リストを印刷する」ページから
r = requests.get("https://www.amazon.co.jp/hz/wishlist/printview/CML0A09KB3DJ", timeout=(10, 10))
soup = BeautifulSoup(r.content, "html.parser")

#soupがうまくいっていなかったら、やめる
if len(soup)==0:
    print("接続エラー")
    sys.exit()
else:
    pass

#情報の抜き出し
list_title = []
list_money = []

for i in soup.find_all("tr"):
    title = i.find_all("span")[0].text
    money = i.find_all("span")[2].text
    #リストに追加
    list_title.append(title)
    list_money.append(money)

#表作成
df = pd.DataFrame()
#リストの1つめは空白のため
df["title"] = list_title[1:]
df["money"] = list_money[1:]
#上下ひっくり返す
df_ = df.iloc[::-1]
#表の更新
df_.to_csv("amazon_list.csv")
    
#google books apiを使用
list_authors = []
list_description = []
list_genre = []

#keyword
keyword = ["ロードバイク", "ipad", "Python", "プログラミング"]

for title in list_title[1:]:
    r = requests.get("https://www.googleapis.com/books/v1/volumes?q=" + title + "&maxResults=1")
    json_data = json.loads(r.text)
    
    #得た情報
    try:
        authors = json_data["items"][0]['volumeInfo']["authors"][0]
    except KeyError:
        authors = "nan"
    
    try:
        description = json_data["items"][0]['volumeInfo']["description"]
    except KeyError:
        description = "nan"
    
    #ジャンルの推定
    for key in keyword:
        if key in description:
            genre = key
        else:
            genre = "nan"
            
    #リストに入れ込む
    list_authors.append(authors)
    list_description.append(description)
    list_genre.append(genre)
    
df["authors"] = list_authors
df["genre"] = list_genre
df_ = df.iloc[::-1]

#gssに書き込む
scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
 
#認証情報の設定
credentials = ServiceAccountCredentials.from_json_keyfile_name('./key/code.json', scope)
 
#Googleスプレッドシートの取得
gClient = gspread.authorize(credentials)
SPREADSHEET_KEY = '1O2DWEIZ9QElv4I8djn_hwwW-EBZwGyQHjFIj5ukddPI'
worksheet = gClient.open_by_key(SPREADSHEET_KEY).sheet1

#編集
range_cell = "E" + str(len(df_) + 1)

cell_list = worksheet.range('B2:' + range_cell)
for cell in cell_list:
    val = df_.iloc[cell.row-2][cell.col-2]
    cell.value = val

# スプレッドシートに書き出す
worksheet.update_cells(cell_list)

#終了
print("書き込んだよ！")
```
